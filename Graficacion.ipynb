```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datos=pd.read_csv("/content/casasboston.csv")
```

```python
print(datos)
```

```python
df=datos[["RM","CRIM","MEDV","TOWN","CHAS"]]
```

```python
#Como renombrar columnas de datos...
df=datos.rename(columns={
    "TOWN":"CIUDAD",
    "CRIM":"INDICE_CRIMEN", "INDUS":"PCT_ZONA:INDUSTRIAL",
    "CHAS":"RIO CHARLES", "RM":"N_HABITACIONES_MEDIO",
    "MEDV":"VALOR_MEDIANO", "LSTAT":"PCT_CLASE_BAJA"
})
print(df)
```

```python
df.N_HABITACIONES_MEDIO.plot.hist()
plt.show()
```

```python
df.N_HABITACIONES_MEDIO.plot.hist()
plt.xlabel('Eso Tilin')
plt.ylabel('Wow Tilin')
plt.title('Vaya Tilin')
```

```python
df.N_HABITACIONES_MEDIO.plot.line()

plt.xlabel('Eso Tilin')
plt.ylabel('Wow Tilin')
plt.title('Vaya Tilin')
```

```python
valor_por_ciudad=df.groupby("CIUDAD")["VALOR_MEDIANO"].mean()
valor_por_ciudad.head(10).plot.barh()
plt.show()
```
